```python
import os
import cv2
import numpy as np
from tqdm import tqdm


def create_mask_from_yolo_segmentation(label_path, image_path, output_path):
    image = cv2.imread(image_path)

    if image is None:
        return

    height, width = image.shape[:2]

    mask = np.zeros((height, width), dtype=np.uint8)

    with open(label_path, 'r') as f:
        lines = f.readlines()

    for line in lines:
        values = line.strip().split()

        polygon = np.array(values[1:], dtype=np.float32)
        polygon = polygon.reshape(-1, 2)

        polygon[:, 0] *= width
        polygon[:, 1] *= height

        polygon = polygon.astype(np.int32)

        cv2.fillPoly(mask, [polygon], 255)

    cv2.imwrite(output_path, mask)


def generate_masks(images_dir, labels_dir, output_dir):
    os.makedirs(output_dir, exist_ok=True)

    image_files = os.listdir(images_dir)

    for image_file in tqdm(image_files):

        image_path = os.path.join(images_dir, image_file)

        label_file = os.path.splitext(image_file)[0] + ".txt"
        label_path = os.path.join(labels_dir, label_file)

        output_path = os.path.join(
            output_dir,
            os.path.splitext(image_file)[0] + ".png"
        )

        if os.path.exists(label_path):
            create_mask_from_yolo_segmentation(
                label_path,
                image_path,
                output_path
            )

    print(f"Masks generated successfully in: {output_dir}")
```

```python
dataset_code = """
import os
import cv2
import torch
import numpy as np

from torch.utils.data import Dataset


class PotholeDataset(Dataset):

    def __init__(
        self,
        image_dir,
        mask_dir,
        transform=None
    ):

        self.image_dir = image_dir
        self.mask_dir = mask_dir
        self.transform = transform

        self.images = sorted(os.listdir(image_dir))

    def __len__(self):
        return len(self.images)

    def __getitem__(self, index):

        image_name = self.images[index]

        image_path = os.path.join(
            self.image_dir,
            image_name
        )

        mask_name = os.path.splitext(image_name)[0] + ".png"

        mask_path = os.path.join(
            self.mask_dir,
            mask_name
        )

        # Read image
        image = cv2.imread(image_path)
        image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

        # Read mask
        mask = cv2.imread(mask_path, cv2.IMREAD_GRAYSCALE)

        # Binary mask
        mask = (mask > 0).astype(np.float32)

        if self.transform:

            augmented = self.transform(
                image=image,
                mask=mask
            )

            image = augmented["image"]
            mask = augmented["mask"]

        mask = torch.unsqueeze(mask, dim=0)

        return image, mask
"""

with open("ARSFormer/src/data/dataset.py", "w") as f:
    f.write(dataset_code)

print("dataset.py created successfully!")
```

```python
loss_code = """
import torch
import torch.nn as nn


class DiceLoss(nn.Module):

    def __init__(self, smooth=1e-6):
        super().__init__()
        self.smooth = smooth

    def forward(self, preds, targets):

        preds = torch.sigmoid(preds)

        preds = preds.view(-1)
        targets = targets.view(-1)

        intersection = (preds * targets).sum()

        dice = (
            2.0 * intersection + self.smooth
        ) / (
            preds.sum() + targets.sum() + self.smooth
        )

        return 1 - dice


class HybridLoss(nn.Module):

    def __init__(self):

        super().__init__()

        self.dice = DiceLoss()

        self.bce = nn.BCEWithLogitsLoss()

    def forward(self, preds, targets):

        dice_loss = self.dice(preds, targets)

        bce_loss = self.bce(preds, targets)

        total_loss = (
            0.5 * dice_loss +
            0.5 * bce_loss
        )

        return total_loss
"""

with open("ARSFormer/src/losses/hybrid_loss.py", "w") as f:
    f.write(loss_code)

print("hybrid_loss.py created successfully!")
```

```python
metrics_code = """
import torch


def dice_score(preds, targets, smooth=1e-6):

    preds = torch.sigmoid(preds)
    preds = (preds > 0.5).float()

    preds = preds.view(-1)
    targets = targets.view(-1)

    intersection = (preds * targets).sum()

    dice = (
        2.0 * intersection + smooth
    ) / (
        preds.sum() + targets.sum() + smooth
    )

    return dice.item()


def iou_score(preds, targets, smooth=1e-6):

    preds = torch.sigmoid(preds)
    preds = (preds > 0.5).float()

    preds = preds.view(-1)
    targets = targets.view(-1)

    intersection = (preds * targets).sum()

    union = (
        preds.sum() +
        targets.sum() -
        intersection
    )

    iou = (
        intersection + smooth
    ) / (
        union + smooth
    )

    return iou.item()
"""

with open("ARSFormer/src/training/metrics.py", "w") as f:
    f.write(metrics_code)

print("metrics.py created successfully!")
```

```python
trainer_code = """
import torch
from tqdm import tqdm

from src.training.metrics import (
    dice_score,
    iou_score
)


def train_one_epoch(
    model,
    loader,
    optimizer,
    criterion,
    device,
    scaler
):

    model.train()

    train_loss = 0

    for images, masks in tqdm(loader):

        images = images.to(device)
        masks = masks.to(device)

        optimizer.zero_grad()

        with torch.cuda.amp.autocast():

            outputs = model(images)

            loss = criterion(outputs, masks)

        scaler.scale(loss).backward()

        scaler.step(optimizer)

        scaler.update()

        train_loss += loss.item()

    return train_loss / len(loader)


def validate(
    model,
    loader,
    criterion,
    device
):

    model.eval()

    val_loss = 0

    dice_total = 0
    iou_total = 0

    with torch.no_grad():

        for images, masks in tqdm(loader):

            images = images.to(device)
            masks = masks.to(device)

            outputs = model(images)

            loss = criterion(outputs, masks)

            val_loss += loss.item()

            dice_total += dice_score(outputs, masks)

            iou_total += iou_score(outputs, masks)

    avg_loss = val_loss / len(loader)

    avg_dice = dice_total / len(loader)

    avg_iou = iou_total / len(loader)

    return avg_loss, avg_dice, avg_iou
"""

with open("ARSFormer/src/training/trainer.py", "w") as f:
    f.write(trainer_code)

print("trainer.py created successfully!")
```

```python
cbam_code = """
import torch
import torch.nn as nn


class ChannelAttention(nn.Module):

    def __init__(self, in_channels, reduction=16):

        super().__init__()

        self.avg_pool = nn.AdaptiveAvgPool2d(1)
        self.max_pool = nn.AdaptiveMaxPool2d(1)

        self.fc = nn.Sequential(
            nn.Conv2d(in_channels, in_channels // reduction, 1, bias=False),
            nn.ReLU(),
            nn.Conv2d(in_channels // reduction, in_channels, 1, bias=False)
        )

        self.sigmoid = nn.Sigmoid()

    def forward(self, x):

        avg_out = self.fc(self.avg_pool(x))
        max_out = self.fc(self.max_pool(x))

        out = avg_out + max_out

        return self.sigmoid(out)


class SpatialAttention(nn.Module):

    def __init__(self, kernel_size=7):

        super().__init__()

        self.conv = nn.Conv2d(
            2,
            1,
            kernel_size,
            padding=kernel_size // 2,
            bias=False
        )

        self.sigmoid = nn.Sigmoid()

    def forward(self, x):

        avg_out = torch.mean(x, dim=1, keepdim=True)

        max_out, _ = torch.max(x, dim=1, keepdim=True)

        x = torch.cat([avg_out, max_out], dim=1)

        x = self.conv(x)

        return self.sigmoid(x)


class CBAM(nn.Module):

    def __init__(self, in_channels):

        super().__init__()

        self.channel_attention = ChannelAttention(in_channels)

        self.spatial_attention = SpatialAttention()

    def forward(self, x):

        x = x * self.channel_attention(x)

        x = x * self.spatial_attention(x)

        return x
"""

with open("ARSFormer/src/models/cbam.py", "w") as f:
    f.write(cbam_code)

print("cbam.py created successfully!")
```

```python
edge_code = """
import torch
import torch.nn as nn


class EdgeRefinementModule(nn.Module):

    def __init__(self, in_channels):

        super().__init__()

        self.refine = nn.Sequential(

            nn.Conv2d(
                in_channels,
                in_channels,
                kernel_size=3,
                padding=1,
                bias=False
            ),

            nn.BatchNorm2d(in_channels),

            nn.ReLU(inplace=True),

            nn.Conv2d(
                in_channels,
                in_channels,
                kernel_size=3,
                padding=1,
                bias=False
            ),

            nn.BatchNorm2d(in_channels),

            nn.ReLU(inplace=True)
        )

    def forward(self, x):

        refined = self.refine(x)

        return refined + x
"""

with open("ARSFormer/src/models/edge_refinement.py", "w") as f:
    f.write(edge_code)

print("edge_refinement.py created successfully!")
```

```python
arsformer_code = """
import torch
import torch.nn as nn
import torch.nn.functional as F

from transformers import SegformerForSemanticSegmentation

from src.models.cbam import CBAM
from src.models.edge_refinement import EdgeRefinementModule


class ARSFormer(nn.Module):

    def __init__(self, num_classes=1):

        super().__init__()

        self.backbone = SegformerForSemanticSegmentation.from_pretrained(
            "nvidia/segformer-b5-finetuned-ade-640-640",
            num_labels=num_classes,
            ignore_mismatched_sizes=True
        )

        # Attention Refinement
        self.cbam = CBAM(1)

        # Edge Refinement
        self.edge_refinement = EdgeRefinementModule(1)

    def forward(self, x):

        outputs = self.backbone(pixel_values=x)

        logits = outputs.logits

        logits = F.interpolate(
            logits,
            size=x.shape[2:],
            mode="bilinear",
            align_corners=False
        )

        # CBAM Attention
        logits = self.cbam(logits)

        # Edge Refinement
        logits = self.edge_refinement(logits)

        return logits
"""

with open("ARSFormer/src/models/arsformer.py", "w") as f:
    f.write(arsformer_code)

print("arsformer.py created successfully!")
```

```python
cbam_simple = """
import torch
import torch.nn as nn


class SpatialAttention(nn.Module):

    def __init__(self, kernel_size=7):

        super().__init__()

        self.conv = nn.Conv2d(
            2,
            1,
            kernel_size,
            padding=kernel_size // 2,
            bias=False
        )

        self.sigmoid = nn.Sigmoid()

    def forward(self, x):

        avg_out = torch.mean(
            x,
            dim=1,
            keepdim=True
        )

        max_out, _ = torch.max(
            x,
            dim=1,
            keepdim=True
        )

        attention = torch.cat(
            [avg_out, max_out],
            dim=1
        )

        attention = self.conv(attention)

        attention = self.sigmoid(attention)

        return x * attention


class CBAM(nn.Module):

    def __init__(self, in_channels):

        super().__init__()

        self.spatial_attention = SpatialAttention()

    def forward(self, x):

        x = self.spatial_attention(x)

        return x
"""

with open("ARSFormer/src/models/cbam.py", "w") as f:
    f.write(cbam_simple)

print("Simplified CBAM saved successfully!")
```

```python
improved_loss = """
import torch
import torch.nn as nn
import torch.nn.functional as F


class DiceLoss(nn.Module):

    def __init__(self, smooth=1e-6):

        super().__init__()

        self.smooth = smooth

    def forward(self, preds, targets):

        preds = torch.sigmoid(preds)

        preds = preds.view(-1)
        targets = targets.view(-1)

        intersection = (preds * targets).sum()

        dice = (
            2 * intersection + self.smooth
        ) / (
            preds.sum() + targets.sum() + self.smooth
        )

        return 1 - dice


class FocalLoss(nn.Module):

    def __init__(self, alpha=0.8, gamma=2):

        super().__init__()

        self.alpha = alpha
        self.gamma = gamma

    def forward(self, preds, targets):

        bce = F.binary_cross_entropy_with_logits(
            preds,
            targets,
            reduction='none'
        )

        probs = torch.sigmoid(preds)

        focal = (
            self.alpha *
            (1 - probs) ** self.gamma *
            bce
        )

        return focal.mean()


class ImprovedHybridLoss(nn.Module):

    def __init__(self):

        super().__init__()

        self.dice = DiceLoss()

        self.focal = FocalLoss()

    def forward(self, preds, targets):

        dice_loss = self.dice(preds, targets)

        focal_loss = self.focal(preds, targets)

        total_loss = (
            0.6 * dice_loss +
            0.4 * focal_loss
        )

        return total_loss
"""

with open("ARSFormer/src/losses/improved_hybrid_loss.py", "w") as f:
    f.write(improved_loss)

print("Improved hybrid loss created successfully!")
```

```python
boundary_loss_code = """
import torch
import torch.nn as nn
import torch.nn.functional as F


class DiceLoss(nn.Module):

    def __init__(self, smooth=1e-6):

        super().__init__()

        self.smooth = smooth

    def forward(self, preds, targets):

        preds = torch.sigmoid(preds)

        preds = preds.view(-1)
        targets = targets.view(-1)

        intersection = (preds * targets).sum()

        dice = (
            2 * intersection + self.smooth
        ) / (
            preds.sum() + targets.sum() + self.smooth
        )

        return 1 - dice


class FocalLoss(nn.Module):

    def __init__(self, alpha=0.8, gamma=2):

        super().__init__()

        self.alpha = alpha
        self.gamma = gamma

    def forward(self, preds, targets):

        bce = F.binary_cross_entropy_with_logits(
            preds,
            targets,
            reduction='none'
        )

        probs = torch.sigmoid(preds)

        focal = (
            self.alpha *
            (1 - probs) ** self.gamma *
            bce
        )

        return focal.mean()


class BoundaryLoss(nn.Module):

    def __init__(self):

        super().__init__()

    def forward(self, preds, targets):

        preds = torch.sigmoid(preds)

        # Sobel edge filters
        sobel_x = torch.tensor(
            [[-1,0,1],
             [-2,0,2],
             [-1,0,1]],
            dtype=torch.float32,
            device=preds.device
        ).view(1,1,3,3)

        sobel_y = torch.tensor(
            [[-1,-2,-1],
             [0,0,0],
             [1,2,1]],
            dtype=torch.float32,
            device=preds.device
        ).view(1,1,3,3)

        pred_edges_x = F.conv2d(
            preds,
            sobel_x,
            padding=1
        )

        pred_edges_y = F.conv2d(
            preds,
            sobel_y,
            padding=1
        )

        target_edges_x = F.conv2d(
            targets,
            sobel_x,
            padding=1
        )

        target_edges_y = F.conv2d(
            targets,
            sobel_y,
            padding=1
        )

        pred_edges = torch.sqrt(
            pred_edges_x**2 +
            pred_edges_y**2
        )

        target_edges = torch.sqrt(
            target_edges_x**2 +
            target_edges_y**2
        )

        return F.l1_loss(
            pred_edges,
            target_edges
        )


class BoundaryAwareLoss(nn.Module):

    def __init__(self):

        super().__init__()

        self.dice = DiceLoss()
        self.focal = FocalLoss()
        self.boundary = BoundaryLoss()

    def forward(self, preds, targets):

        dice_loss = self.dice(preds, targets)

        focal_loss = self.focal(preds, targets)

        boundary_loss = self.boundary(preds, targets)

        total_loss = (
            0.5 * dice_loss +
            0.3 * focal_loss +
            0.2 * boundary_loss
        )

        return total_loss
"""

with open(
    "ARSFormer/src/losses/boundary_aware_loss.py",
    "w"
) as f:

    f.write(boundary_loss_code)

print("Boundary-aware loss created successfully!")
```

```python
#!/usr/bin/env python3
"""
ARSFormer v3 — SegFormer-B2 + Enhanced FPN Decoder
=====================================================
Problem with v1/v2: SegFormer's default MLP decoder is too weak.
It merges multi-scale features with simple linear projections,
losing spatial detail critical for pothole boundary accuracy.

Fix: Replace MLP decoder with a Feature Pyramid Network (FPN) decoder
that properly fuses multi-scale transformer features with skip connections.

Expected improvement: +2-4% IoU over v2 (0.8383 → 0.86+)
"""

import os
import sys
import cv2
import numpy as np
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import Dataset, DataLoader
from torch.cuda.amp import GradScaler, autocast
import albumentations as A
from albumentations.pytorch import ToTensorV2
from glob import glob
from tqdm import tqdm
import matplotlib.pyplot as plt
from transformers import SegformerModel, SegformerConfig

# ── Config ───────────────────────────────────────────────────
class CFG:
    TRAIN_IMG_DIR  = "ARSFormer/final_dataset/train/images"
    TRAIN_MASK_DIR = "ARSFormer/final_dataset/train/masks"
    VALID_IMG_DIR  = "ARSFormer/final_dataset/valid/images"
    VALID_MASK_DIR = "ARSFormer/final_dataset/valid/masks"
    SAVE_DIR       = "ARSFormer/checkpoints"

    IMG_SIZE       = 512
    NUM_CLASSES    = 1
    EPOCHS         = 60
    BATCH_SIZE     = 8
    NUM_WORKERS    = 4
    LR             = 4e-5        # slightly lower for new decoder
    WEIGHT_DECAY   = 1e-4
    SEED           = 42
    AMP            = True
    DEVICE         = torch.device("cuda" if torch.cuda.is_available() else "cpu")

    W_DICE         = 0.45
    W_BCE          = 0.45
    W_BOUNDARY     = 0.10


os.makedirs(CFG.SAVE_DIR, exist_ok=True)
torch.manual_seed(CFG.SEED)
np.random.seed(CFG.SEED)


# ── Enhanced FPN Decoder ─────────────────────────────────────
class ConvBNReLU(nn.Sequential):
    """Basic Conv → BN → ReLU block."""
    def __init__(self, in_ch, out_ch, kernel=3, stride=1, padding=1):
        super().__init__(
            nn.Conv2d(in_ch, out_ch, kernel, stride=stride,
                      padding=padding, bias=False),
            nn.BatchNorm2d(out_ch),
            nn.ReLU(inplace=True)
        )


class FPNDecoder(nn.Module):
    """
    Feature Pyramid Network decoder for SegFormer-B2.

    SegFormer-B2 encoder outputs 4 feature maps at strides 4,8,16,32:
        C1: (B, 64,  H/4,  W/4)
        C2: (B, 128, H/8,  W/8)
        C3: (B, 320, H/16, W/16)
        C4: (B, 512, H/32, W/32)

    FPN:
        1. Project each scale to 256 channels
        2. Top-down pathway with upsampling + addition
        3. Final conv per level → upsample all to H/4 → concat → predict
    """
    def __init__(self,
                 in_channels=(64, 128, 320, 512),
                 fpn_channels=256,
                 num_classes=1):
        super().__init__()

        # Lateral projections (1×1 conv) — one per encoder scale
        self.lateral = nn.ModuleList([
            nn.Sequential(
                nn.Conv2d(in_ch, fpn_channels, 1, bias=False),
                nn.BatchNorm2d(fpn_channels),
                nn.ReLU(inplace=True)
            )
            for in_ch in in_channels
        ])

        # Output convolutions after top-down merge
        self.output_convs = nn.ModuleList([
            ConvBNReLU(fpn_channels, fpn_channels)
            for _ in in_channels
        ])

        # Final head: concat all levels (upsampled to H/4) → predict
        self.head = nn.Sequential(
            ConvBNReLU(fpn_channels * len(in_channels), fpn_channels * 2),
            nn.Dropout2d(0.1),
            ConvBNReLU(fpn_channels * 2, fpn_channels),
            nn.Conv2d(fpn_channels, num_classes, 1)
        )

    def forward(self, features):
        """
        features: list of 4 tensors [C1, C2, C3, C4]
        C1 = finest (H/4), C4 = coarsest (H/32)
        """
        # Step 1: lateral projections
        laterals = [lat(f) for lat, f in zip(self.lateral, features)]

        # Step 2: top-down pathway (coarse → fine)
        # Start from C4, progressively add upsampled coarser features
        for i in range(len(laterals) - 1, 0, -1):
            up = F.interpolate(laterals[i],
                               size=laterals[i - 1].shape[-2:],
                               mode="bilinear", align_corners=False)
            laterals[i - 1] = laterals[i - 1] + up

        # Step 3: output conv at each level
        outs = [conv(lat) for conv, lat in
                zip(self.output_convs, laterals)]

        # Step 4: upsample all levels to finest scale (H/4)
        target_size = outs[0].shape[-2:]
        outs_up = [
            F.interpolate(o, size=target_size,
                          mode="bilinear", align_corners=False)
            for o in outs
        ]

        # Step 5: concat + predict
        fused = torch.cat(outs_up, dim=1)   # (B, 256*4, H/4, W/4)
        return self.head(fused)              # (B, 1, H/4, W/4)


# ── Full ARSFormer-v3 Model ───────────────────────────────────
class ARSFormerV3(nn.Module):
    """
    ARSFormer v3:
        Encoder : SegFormer-B2 (MiT-B2 transformer, pretrained on ADE20K)
        Decoder : Custom FPN with top-down feature pyramid
        Output  : Raw logits (B, 1, H, W)

    Key difference from v1/v2:
        v1/v2 used SegFormer's built-in MLP decoder (too simple)
        v3 replaces it with FPN — proper multi-scale fusion with skip connections
    """
    # SegFormer-B2 encoder output channels per stage
    ENCODER_CHANNELS = (64, 128, 320, 512)

    def __init__(self, num_classes=1, fpn_channels=256):
        super().__init__()

        # Load only the encoder (not the full segmentation model)
        self.encoder = SegformerModel.from_pretrained(
            "nvidia/mit-b2",
            output_hidden_states=True,
        )

        self.decoder = FPNDecoder(
            in_channels=self.ENCODER_CHANNELS,
            fpn_channels=fpn_channels,
            num_classes=num_classes
        )

    def forward(self, x):
        input_size = x.shape[-2:]

        # Encoder — get all 4 hidden states
        outputs       = self.encoder(
            pixel_values=x,
            output_hidden_states=True,
            return_dict=True
        )
        # hidden_states: tuple of 4 tensors, finest to coarsest
        hidden_states = outputs.hidden_states   # (C1, C2, C3, C4)

        # Decoder
        logits = self.decoder(list(hidden_states))   # (B, 1, H/4, W/4)

        # Upsample to input resolution
        logits = F.interpolate(logits, size=input_size,
                               mode="bilinear", align_corners=False)
        return logits

    def get_parameter_count(self):
        total     = sum(p.numel() for p in self.parameters()) / 1e6
        trainable = sum(p.numel() for p in self.parameters()
                        if p.requires_grad) / 1e6
        return total, trainable

    def get_param_groups(self, lr):
        """Differential LR: encoder gets 0.1×, decoder gets full LR."""
        return [
            {"params": self.encoder.parameters(), "lr": lr * 0.1},
            {"params": self.decoder.parameters(), "lr": lr},
        ]


# ── Loss Functions ───────────────────────────────────────────
class DiceLoss(nn.Module):
    def __init__(self, smooth=1.0):
        super().__init__()
        self.smooth = smooth

    def forward(self, logits, targets):
        preds = torch.sigmoid(logits).view(-1)
        tgts  = targets.view(-1)
        inter = (preds * tgts).sum()
        return 1 - (2 * inter + self.smooth) / \
               (preds.sum() + tgts.sum() + self.smooth)


class BoundaryLoss(nn.Module):
    def __init__(self):
        super().__init__()
        sobel_x = torch.tensor(
            [[-1, 0, 1], [-2, 0, 2], [-1, 0, 1]],
            dtype=torch.float32).view(1, 1, 3, 3)
        sobel_y = torch.tensor(
            [[-1, -2, -1], [0, 0, 0], [1, 2, 1]],
            dtype=torch.float32).view(1, 1, 3, 3)
        self.register_buffer("sobel_x", sobel_x)
        self.register_buffer("sobel_y", sobel_y)

    def get_boundary(self, mask):
        ex  = F.conv2d(mask, self.sobel_x, padding=1)
        ey  = F.conv2d(mask, self.sobel_y, padding=1)
        bnd = torch.sqrt(ex ** 2 + ey ** 2 + 1e-6)
        return bnd / (bnd.max() + 1e-6)

    def forward(self, logits, targets):
        pred_probs    = torch.sigmoid(logits)
        gt_boundary   = self.get_boundary(targets)
        pred_boundary = self.get_boundary(pred_probs)
        return F.mse_loss(pred_boundary, gt_boundary)


class HybridLoss(nn.Module):
    def __init__(self, w_dice=0.45, w_bce=0.45, w_boundary=0.10):
        super().__init__()
        self.w_dice     = w_dice
        self.w_bce      = w_bce
        self.w_boundary = w_boundary
        self.dice       = DiceLoss()
        self.bce        = nn.BCEWithLogitsLoss()
        self.boundary   = BoundaryLoss()

    def forward(self, logits, targets):
        l_dice     = self.dice(logits, targets)
        l_bce      = self.bce(logits, targets)
        l_boundary = self.boundary(logits, targets)
        total      = (self.w_dice     * l_dice +
                      self.w_bce      * l_bce  +
                      self.w_boundary * l_boundary)
        return total, l_dice, l_bce, l_boundary


# ── Dataset ──────────────────────────────────────────────────
class PotholeDataset(Dataset):
    def __init__(self, img_dir, mask_dir, transform=None):
        self.transform   = transform
        self.image_paths = []
        for ext in ["*.jpg", "*.jpeg", "*.png",
                    "*.JPG", "*.JPEG", "*.PNG"]:
            self.image_paths.extend(glob(os.path.join(img_dir, ext)))
        self.image_paths = sorted(set(self.image_paths))

        self.mask_paths = []
        valid_pairs     = []
        for ip in self.image_paths:
            stem      = os.path.splitext(os.path.basename(ip))[0]
            mask_path = os.path.join(mask_dir, stem + ".png")
            if os.path.exists(mask_path):
                self.mask_paths.append(mask_path)
                valid_pairs.append(ip)
        self.image_paths = valid_pairs

    def __len__(self):
        return len(self.image_paths)

    def __getitem__(self, idx):
        image = cv2.imread(self.image_paths[idx])
        image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
        mask  = cv2.imread(self.mask_paths[idx], cv2.IMREAD_GRAYSCALE)
        mask  = (mask > 127).astype(np.float32)
        if self.transform:
            aug   = self.transform(image=image, mask=mask)
            image = aug["image"]
            mask  = aug["mask"].unsqueeze(0)
        return image, mask


def get_train_transforms():
    return A.Compose([
        A.RandomResizedCrop(
            size=(CFG.IMG_SIZE, CFG.IMG_SIZE),
            scale=(0.6, 1.0), p=1.0),
        A.HorizontalFlip(p=0.5),
        A.VerticalFlip(p=0.2),
        A.ShiftScaleRotate(shift_limit=0.1, scale_limit=0.2,
                           rotate_limit=20, p=0.5),
        A.OneOf([
            A.RandomFog(fog_coef_lower=0.1, fog_coef_upper=0.3, p=1.0),
            A.RandomShadow(num_shadows_lower=1,
                           num_shadows_upper=2, p=1.0),
            A.MotionBlur(blur_limit=7, p=1.0),
            A.GaussNoise(var_limit=(20, 80), p=1.0),
        ], p=0.5),
        A.OneOf([
            A.RandomBrightnessContrast(brightness_limit=0.3,
                                       contrast_limit=0.3, p=1.0),
            A.CLAHE(clip_limit=4.0, p=1.0),
            A.HueSaturationValue(hue_shift_limit=10,
                                 sat_shift_limit=20,
                                 val_shift_limit=20, p=1.0),
        ], p=0.5),
        A.Normalize(mean=(0.485, 0.456, 0.406),
                    std=(0.229, 0.224, 0.225)),
        ToTensorV2(),
    ])


def get_val_transforms():
    return A.Compose([
        A.Resize(CFG.IMG_SIZE, CFG.IMG_SIZE),
        A.Normalize(mean=(0.485, 0.456, 0.406),
                    std=(0.229, 0.224, 0.225)),
        ToTensorV2(),
    ])


# ── Metrics ──────────────────────────────────────────────────
class SegMetrics:
    def __init__(self, thresh=0.5, eps=1e-6):
        self.thresh = thresh
        self.eps    = eps
        self.reset()

    def reset(self):
        self.tp = self.fp = self.fn = 0.0

    @torch.no_grad()
    def update(self, logits, targets):
        preds    = (torch.sigmoid(logits) > self.thresh).float()
        self.tp += (preds * targets).sum().item()
        self.fp += (preds * (1 - targets)).sum().item()
        self.fn += ((1 - preds) * targets).sum().item()

    def iou(self):
        return self.tp / (self.tp + self.fp + self.fn + self.eps)

    def dice(self):
        return (2 * self.tp) / (2 * self.tp + self.fp + self.fn + self.eps)


# ── TTA ──────────────────────────────────────────────────────
@torch.no_grad()
def tta_predict(model, images, amp=True):
    preds = []
    with autocast(enabled=amp):
        preds.append(torch.sigmoid(model(images)))

    for dims in [[3], [2], [2, 3]]:
        flipped = torch.flip(images, dims=dims)
        with autocast(enabled=amp):
            pred = torch.sigmoid(model(flipped))
        preds.append(torch.flip(pred, dims=dims))

    return torch.stack(preds).mean(dim=0)


# ── Train / Val ──────────────────────────────────────────────
def train_one_epoch(model, loader, optimizer, scheduler, criterion, scaler):
    model.train()
    metrics    = SegMetrics()
    total_loss = d_l = b_l = bnd_l = 0.0

    pbar = tqdm(loader, desc="  Train", leave=False)
    for images, masks in pbar:
        images = images.to(CFG.DEVICE, non_blocking=True)
        masks  = masks.to(CFG.DEVICE,  non_blocking=True)

        optimizer.zero_grad(set_to_none=True)
        with autocast(enabled=CFG.AMP):
            logits            = model(images)
            loss, ld, lb, lbn = criterion(logits, masks)

        scaler.scale(loss).backward()
        scaler.unscale_(optimizer)
        torch.nn.utils.clip_grad_norm_(model.parameters(), 1.0)
        scaler.step(optimizer)
        scaler.update()
        scheduler.step()

        metrics.update(logits.detach(), masks)
        total_loss += loss.item()
        d_l        += ld.item()
        b_l        += lb.item()
        bnd_l      += lbn.item()

        pbar.set_postfix(loss=f"{loss.item():.4f}",
                         iou=f"{metrics.iou():.4f}",
                         bnd=f"{lbn.item():.4f}")

    n = len(loader)
    return total_loss/n, d_l/n, b_l/n, bnd_l/n, metrics.iou(), metrics.dice()


@torch.no_grad()
def validate(model, loader, criterion, use_tta=False):
    model.eval()
    metrics    = SegMetrics()
    total_loss = 0.0

    for images, masks in tqdm(loader, desc="  Val  ", leave=False):
        images = images.to(CFG.DEVICE, non_blocking=True)
        masks  = masks.to(CFG.DEVICE,  non_blocking=True)

        if use_tta:
            probs         = tta_predict(model, images, amp=CFG.AMP)
            probs_clamped = probs.clamp(1e-6, 1 - 1e-6)
            logits        = torch.log(probs_clamped / (1 - probs_clamped))
        else:
            with autocast(enabled=CFG.AMP):
                logits = model(images)

        loss, *_ = criterion(logits, masks)
        metrics.update(logits, masks)
        total_loss += loss.item()

    return total_loss / len(loader), metrics.iou(), metrics.dice()


# ── Main ─────────────────────────────────────────────────────
def main():
    print(f"Device  : {CFG.DEVICE}")
    print(f"GPU     : {torch.cuda.get_device_name(0)}")
    print(f"Model   : ARSFormer-v3 (MiT-B2 + FPN Decoder)")
    print(f"Loss    : Dice×{CFG.W_DICE} + BCE×{CFG.W_BCE} "
          f"+ Boundary×{CFG.W_BOUNDARY}")

    train_ds = PotholeDataset(CFG.TRAIN_IMG_DIR, CFG.TRAIN_MASK_DIR,
                               get_train_transforms())
    val_ds   = PotholeDataset(CFG.VALID_IMG_DIR, CFG.VALID_MASK_DIR,
                               get_val_transforms())
    print(f"Train   : {len(train_ds)}  |  Val : {len(val_ds)}")

    train_loader = DataLoader(
        train_ds, batch_size=CFG.BATCH_SIZE, shuffle=True,
        num_workers=CFG.NUM_WORKERS, pin_memory=True, drop_last=True)
    val_loader   = DataLoader(
        val_ds, batch_size=CFG.BATCH_SIZE, shuffle=False,
        num_workers=CFG.NUM_WORKERS, pin_memory=True)

    # Model
    model = ARSFormerV3(num_classes=CFG.NUM_CLASSES).to(CFG.DEVICE)
    total, trainable = model.get_parameter_count()
    print(f"Params  : {total:.1f}M total  |  {trainable:.1f}M trainable")

    # Sanity check
    with torch.no_grad():
        dummy = torch.randn(2, 3, 512, 512).to(CFG.DEVICE)
        out   = model(dummy)
        print(f"Output  : {out.shape}")   # expect (2, 1, 512, 512)

    criterion = HybridLoss(CFG.W_DICE, CFG.W_BCE, CFG.W_BOUNDARY).to(CFG.DEVICE)

    optimizer = torch.optim.AdamW(
        model.get_param_groups(CFG.LR),
        weight_decay=CFG.WEIGHT_DECAY
    )

    scheduler = torch.optim.lr_scheduler.OneCycleLR(
        optimizer,
        max_lr=[CFG.LR * 0.1, CFG.LR],
        epochs=CFG.EPOCHS,
        steps_per_epoch=len(train_loader),
        pct_start=0.1,
        anneal_strategy="cos",
        div_factor=10.0,
        final_div_factor=100.0,
    )

    scaler   = GradScaler(enabled=CFG.AMP)
    history  = {"train_iou": [], "val_iou": [],
                "train_dice": [], "val_dice": [],
                "train_loss": [], "val_loss": [],
                "boundary_loss": []}
    best_iou  = 0.0
    best_path = os.path.join(
    CFG.SAVE_DIR,
    "experiment_try_1.pth"
)

    for epoch in range(1, CFG.EPOCHS + 1):
        print(f"\nEpoch [{epoch:02d}/{CFG.EPOCHS}]")

        tr = train_one_epoch(model, train_loader, optimizer,
                             scheduler, criterion, scaler)
        tr_loss, tr_d, tr_b, tr_bnd, tr_iou, tr_dice = tr

        use_tta = (epoch >= CFG.EPOCHS - 9)
        va_loss, va_iou, va_dice = validate(
            model, val_loader, criterion, use_tta=use_tta)

        history["train_iou"].append(tr_iou)
        history["val_iou"].append(va_iou)
        history["train_dice"].append(tr_dice)
        history["val_dice"].append(va_dice)
        history["train_loss"].append(tr_loss)
        history["val_loss"].append(va_loss)
        history["boundary_loss"].append(tr_bnd)

        tta_tag = " [TTA]" if use_tta else ""
        print(f"  Train → loss={tr_loss:.4f}  IoU={tr_iou:.4f}  "
              f"Dice={tr_dice:.4f}  BndLoss={tr_bnd:.4f}")
        print(f"  Val{tta_tag}   → loss={va_loss:.4f}  "
              f"IoU={va_iou:.4f}  Dice={va_dice:.4f}")

        if va_iou > best_iou:
            best_iou = va_iou
            torch.save({
                "epoch"      : epoch,
                "model_state": model.state_dict(),
                "val_iou"    : best_iou,
                "val_dice"   : va_dice,
                "model"      : "ARSFormer-v3",
                "decoder"    : "FPN",
                "encoder"    : "MiT-B2",
            }, best_path)
            print(f"  ✅ Best IoU: {best_iou:.4f} — saved")

    print(f"\n🏁 Training complete.")
    print(f"Best Val IoU  : {best_iou:.4f}")

    # Final eval with full TTA
    print("\nFinal evaluation with full TTA on best checkpoint...")
    ckpt = torch.load(best_path, map_location=CFG.DEVICE)
    model.load_state_dict(ckpt["model_state"], strict=False)
    _, final_iou, final_dice = validate(
        model, val_loader, criterion, use_tta=True)
    print(f"Final TTA IoU  : {final_iou:.4f}  ({final_iou*100:.2f}%)")
    print(f"Final TTA Dice : {final_dice:.4f}  ({final_dice*100:.2f}%)")

    # Summary table
    print("\n" + "="*50)
    print("  ARSFormer Version Comparison")
    print("="*50)
    print(f"  v1 (B2 + MLP decoder)        IoU: 0.8343  Dice: 0.9097")
    print(f"  v2 (B2 + boundary + TTA)     IoU: 0.8383  Dice: 0.9121")
    print(f"  v3 (B2 + FPN decoder + TTA)  IoU: {final_iou:.4f}  "
          f"Dice: {final_dice:.4f}")
    print("="*50)

    # Curves
    epochs_r = range(1, CFG.EPOCHS + 1)
    fig, axes = plt.subplots(1, 4, figsize=(22, 5))
    fig.suptitle("ARSFormer-v3 (MiT-B2 + FPN Decoder) — Training Curves",
                 fontsize=13)

    axes[0].plot(epochs_r, history["train_iou"],  label="Train")
    axes[0].plot(epochs_r, history["val_iou"],    label="Val")
    axes[0].axhline(0.8383, color="gray", linestyle="--",
                    label="v2 best (0.8383)")
    axes[0].axhline(0.8500, color="green", linestyle=":",
                    label="Target (0.85)")
    axes[0].set_title("IoU")
    axes[0].legend(); axes[0].grid(alpha=0.3)

    axes[1].plot(epochs_r, history["train_dice"], label="Train")
    axes[1].plot(epochs_r, history["val_dice"],   label="Val")
    axes[1].axhline(0.9121, color="gray", linestyle="--",
                    label="v2 best (0.9121)")
    axes[1].set_title("Dice")
    axes[1].legend(); axes[1].grid(alpha=0.3)

    axes[2].plot(epochs_r, history["train_loss"], label="Train")
    axes[2].plot(epochs_r, history["val_loss"],   label="Val")
    axes[2].set_title("Total Loss")
    axes[2].legend(); axes[2].grid(alpha=0.3)

    axes[3].plot(epochs_r, history["boundary_loss"],
                 color="coral", label="Boundary Loss")
    axes[3].set_title("Boundary Loss (train)")
    axes[3].legend(); axes[3].grid(alpha=0.3)

    plt.tight_layout()
    plt.savefig(os.path.join(CFG.SAVE_DIR, "training_curves_v3.png"), dpi=150)
    plt.show()
    print("Curves saved.")


if __name__ == "__main__":
    main()
```

```python
import os, cv2, warnings, io
import numpy as np
import torch
import torch.nn as nn
import torch.nn.functional as F
import albumentations as A
from albumentations.pytorch import ToTensorV2
import matplotlib
matplotlib.use("Agg")
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
from matplotlib.patches import FancyBboxPatch
from transformers import SegformerModel
from IPython.display import display, clear_output
import ipywidgets as widgets
warnings.filterwarnings("ignore")

DEVICE     = torch.device("cuda" if torch.cuda.is_available() else "cpu")
CKPT_PATH  = "ARSFormer/checkpoints/ARSFormer_v3_FINAL_8667IoU_9286Dice.pth"
IMG_SIZE   = 512
THRESHOLD  = 0.5

# ── Severity config ───────────────────────────────────────────────
SEV_MINOR    = 2.0    # < 2%  pixels → Minor
SEV_MODERATE = 10.0   # 2-10% pixels → Moderate
                      # > 10% pixels → Severe

SEV_NAMES  = {0: "Minor",    1: "Moderate",  2: "Severe"}
SEV_COLORS = {
    0: {"bar": "#43A047", "text": "#43A047"},   # green
    1: {"bar": "#FFB300", "text": "#FFB300"},   # amber
    2: {"bar": "#E53935", "text": "#E53935"},   # red
}
SEV_ACTIONS = {
    0: "Low priority — schedule routine inspection",
    1: "Plan repair within 2–4 weeks",
    2: "Immediate repair required — safety hazard",
}

def mask_to_severity(mask_np):
    ratio = mask_np.sum() / mask_np.size * 100.0
    if ratio < SEV_MINOR:    return 0, ratio
    elif ratio < SEV_MODERATE: return 1, ratio
    else:                      return 2, ratio

def severity_probs_from_mask(mask_np):
    """
    Compute soft severity probabilities from mask coverage ratio.
    Returns array of shape (3,) summing to 1.
    """
    ratio = mask_np.sum() / mask_np.size * 100.0
    probs = np.zeros(3, dtype=np.float32)
    if ratio < SEV_MINOR:
        # Minor — confidence scales with distance from threshold
        conf = max(0.55, 1.0 - ratio / SEV_MINOR * 0.4)
        probs[0] = conf
        probs[1] = (1 - conf) * 0.7
        probs[2] = (1 - conf) * 0.3
    elif ratio < SEV_MODERATE:
        conf = max(0.55, 0.5 + abs(ratio - (SEV_MINOR + SEV_MODERATE)/2) /
                   ((SEV_MODERATE - SEV_MINOR)/2) * 0.4)
        probs[1] = conf
        probs[0] = (1 - conf) * (1 - (ratio - SEV_MINOR) / (SEV_MODERATE - SEV_MINOR))
        probs[2] = (1 - conf) * (ratio - SEV_MINOR) / (SEV_MODERATE - SEV_MINOR)
    else:
        conf = min(0.99, 0.55 + (ratio - SEV_MODERATE) / (100 - SEV_MODERATE) * 0.44)
        probs[2] = conf
        probs[1] = (1 - conf) * 0.7
        probs[0] = (1 - conf) * 0.3
    probs = probs / probs.sum()
    return probs


# ── Model definition ──────────────────────────────────────────────
class ConvBNReLU(nn.Sequential):
    def __init__(self, in_ch, out_ch, kernel=3, stride=1, padding=1):
        super().__init__(
            nn.Conv2d(in_ch, out_ch, kernel, stride=stride,
                      padding=padding, bias=False),
            nn.BatchNorm2d(out_ch),
            nn.ReLU(inplace=True))

class FPNDecoder(nn.Module):
    def __init__(self, in_channels=(64,128,320,512), fpn_channels=256, num_classes=1):
        super().__init__()
        self.lateral = nn.ModuleList([
            nn.Sequential(nn.Conv2d(c, fpn_channels, 1, bias=False),
                          nn.BatchNorm2d(fpn_channels), nn.ReLU(inplace=True))
            for c in in_channels])
        self.output_convs = nn.ModuleList([
            ConvBNReLU(fpn_channels, fpn_channels) for _ in in_channels])
        self.head = nn.Sequential(
            ConvBNReLU(fpn_channels*len(in_channels), fpn_channels*2),
            nn.Dropout2d(0.1),
            ConvBNReLU(fpn_channels*2, fpn_channels),
            nn.Conv2d(fpn_channels, num_classes, 1))

    def forward(self, features):
        laterals = [lat(f) for lat,f in zip(self.lateral, features)]
        for i in range(len(laterals)-1, 0, -1):
            up = F.interpolate(laterals[i], size=laterals[i-1].shape[-2:],
                               mode="bilinear", align_corners=False)
            laterals[i-1] = laterals[i-1] + up
        outs = [conv(lat) for conv,lat in zip(self.output_convs, laterals)]
        target_size = outs[0].shape[-2:]
        outs_up = [F.interpolate(o, size=target_size, mode="bilinear",
                                 align_corners=False) for o in outs]
        return self.head(torch.cat(outs_up, dim=1))

class ARSFormerV3(nn.Module):
    ENCODER_CHANNELS = (64,128,320,512)
    def __init__(self, num_classes=1, fpn_channels=256):
        super().__init__()
        self.encoder = SegformerModel.from_pretrained(
            "nvidia/mit-b2", output_hidden_states=True)
        self.decoder = FPNDecoder(self.ENCODER_CHANNELS, fpn_channels, num_classes)

    def forward(self, x):
        input_size = x.shape[-2:]
        outputs = self.encoder(pixel_values=x, output_hidden_states=True, return_dict=True)
        logits  = self.decoder(list(outputs.hidden_states))
        return F.interpolate(logits, size=input_size, mode="bilinear", align_corners=False)


# ── Load model ────────────────────────────────────────────────────
print("Loading ARSFormer-v3 checkpoint...")
assert os.path.exists(CKPT_PATH), f"Checkpoint not found: {CKPT_PATH}"

ckpt  = torch.load(CKPT_PATH, map_location=DEVICE)
model = ARSFormerV3().to(DEVICE)

state_key = "model_state" if "model_state" in ckpt else "model"
model.load_state_dict(ckpt[state_key])
model.eval()

print(f"✅ Model loaded")
print(f"   Val IoU  : {ckpt.get('val_iou', ckpt.get('iou', '?')):.4f}")
print(f"   Val Dice : {ckpt.get('val_dice', ckpt.get('dice', '?')):.4f}")
print(f"   Epoch    : {ckpt.get('epoch', '?')}")


# ── Transforms ───────────────────────────────────────────────────
transform = A.Compose([
    A.Resize(IMG_SIZE, IMG_SIZE),
    A.Normalize(mean=(0.485,0.456,0.406), std=(0.229,0.224,0.225)),
    ToTensorV2(),
])


# ── TTA predict ───────────────────────────────────────────────────
@torch.no_grad()
def tta_predict(inp):
    preds = []
    for dims in [None, [3], [2], [2,3]]:
        xi = inp.flip(dims) if dims else inp
        logits = model(xi)
        prob   = torch.sigmoid(logits)
        if dims: prob = prob.flip(dims)
        preds.append(prob)
    return torch.stack(preds).mean(0)


# ── Inference ─────────────────────────────────────────────────────
def run_inference(image_bytes):
    # Decode image
    arr     = np.frombuffer(image_bytes, np.uint8)
    img_bgr = cv2.imdecode(arr, cv2.IMREAD_COLOR)
    if img_bgr is None:
        raise ValueError("Cannot decode image. Please upload a valid JPG/PNG.")
    img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)
    h0, w0  = img_rgb.shape[:2]

    # Preprocess
    aug = transform(image=img_rgb, mask=np.zeros((h0,w0), np.uint8))
    inp = aug["image"].unsqueeze(0).to(DEVICE)

    # Inference + TTA
    probs    = tta_predict(inp)
    prob_map = probs[0,0].cpu().numpy()

    # Resize to original
    mask_512  = (prob_map > THRESHOLD).astype(np.uint8)
    mask_orig = cv2.resize(mask_512, (w0,h0), interpolation=cv2.INTER_NEAREST)
    prob_orig = cv2.resize(prob_map, (w0,h0), interpolation=cv2.INTER_LINEAR)

    # Severity from mask area
    sev_id, dmg_pct = mask_to_severity(mask_orig)
    sev_probs       = severity_probs_from_mask(mask_orig)
    sev_conf        = float(sev_probs[sev_id]) * 100

    # Contours
    contours = cv2.findContours(
        mask_orig, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)[0]
    n_holes  = max(1, len(contours)) if mask_orig.sum() > 0 else 0

    # Overlay
    sc      = SEV_COLORS[sev_id]["bar"]
    sc_rgb  = tuple(int(sc.lstrip("#")[i:i+2],16)/255 for i in (0,2,4))
    overlay = img_rgb.astype(float)/255.0
    if mask_orig.sum() > 0:
        for c in range(3):
            overlay[mask_orig==1, c] = (0.45*overlay[mask_orig==1,c] +
                                        0.55*sc_rgb[c])
    overlay = np.clip(overlay, 0, 1)

    # Draw contour borders
    ov_bgr = (overlay*255).astype(np.uint8)
    border_bgr = tuple(int(255*x) for x in reversed(sc_rgb))
    cv2.drawContours(ov_bgr, contours, -1, border_bgr, 2)
    overlay = ov_bgr.astype(float)/255.0

    return {
        "img_rgb":   img_rgb,
        "mask":      mask_orig,
        "prob_map":  prob_orig,
        "overlay":   overlay,
        "sev_id":    sev_id,
        "sev_name":  SEV_NAMES[sev_id],
        "sev_conf":  sev_conf,
        "sev_probs": sev_probs,
        "dmg_pct":   dmg_pct,
        "n_holes":   n_holes,
        "h0": h0, "w0": w0,
    }


# ── Dashboard plot ────────────────────────────────────────────────
def plot_dashboard(r, fname="uploaded image"):
    sev_id    = r["sev_id"]
    sev_name  = r["sev_name"]
    sev_conf  = r["sev_conf"]
    sev_probs = r["sev_probs"]
    dmg_pct   = r["dmg_pct"]
    n_holes   = r["n_holes"]
    sc        = SEV_COLORS[sev_id]["bar"]
    sc_rgb    = tuple(int(sc.lstrip("#")[i:i+2],16)/255 for i in (0,2,4))

    fig = plt.figure(figsize=(20, 11))
    fig.patch.set_facecolor("#0F1117")
    fig.suptitle(
        f"ARSFormer-v3  —  Pothole Analysis  |  {fname}",
        fontsize=14, fontweight="bold", color="white", y=0.97)

    gs = gridspec.GridSpec(2, 4, figure=fig,
                           hspace=0.38, wspace=0.25,
                           top=0.89, bottom=0.05,
                           left=0.03, right=0.97)

    # ── Row 1: image panels ───────────────────────────────────────
    panels = [
        (r["img_rgb"],  "Original image",       None,       None),
        (r["mask"]*255, "Segmentation mask",    "gray",     None),
        (r["prob_map"], "Probability heat map", "RdYlGn_r", [0,1]),
        (r["overlay"],  "Pothole overlay",      None,       None),
    ]
    for j,(data,title,cmap,vr) in enumerate(panels):
        ax = fig.add_subplot(gs[0,j])
        kw = {}
        if cmap: kw["cmap"] = cmap
        if vr:   kw["vmin"],kw["vmax"] = vr
        ax.imshow(data, **kw)
        ax.set_title(title, fontsize=11, color="#CCCCCC", pad=5,
                     fontweight="bold")
        ax.axis("off")
        for sp in ax.spines.values():
            sp.set_edgecolor("#333333"); sp.set_linewidth(0.7)

    # ── Helper: dark panel background ────────────────────────────
    def dark_panel(ax):
        ax.set_facecolor("#1A1D26"); ax.axis("off")
        ax.set_xlim(0,1); ax.set_ylim(0,1)
        ax.add_patch(FancyBboxPatch((0.04,0.04),0.92,0.92,
            boxstyle="round,pad=0.02", facecolor="#1F2330",
            edgecolor="#2A2D3A", linewidth=1.2))

    # ── Panel A: Severity badge ───────────────────────────────────
    axA = fig.add_subplot(gs[1,0])
    dark_panel(axA)
    axA.add_patch(FancyBboxPatch((0.04,0.04),0.92,0.92,
        boxstyle="round,pad=0.02", facecolor="#1F2330",
        edgecolor=sc, linewidth=2))
    axA.text(0.5,0.87,"Severity level", ha="center", fontsize=9,
             color="#888888", transform=axA.transAxes)
    axA.text(0.5,0.63,sev_name.upper(), ha="center", fontsize=28,
             fontweight="bold", color=sc_rgb, transform=axA.transAxes)
    axA.text(0.5,0.44,f"Confidence: {sev_conf:.1f}%", ha="center",
             fontsize=11, color="#CCCCCC", transform=axA.transAxes)
    axA.text(0.5,0.24,SEV_ACTIONS[sev_id], ha="center", fontsize=8,
             color="#888888", style="italic", transform=axA.transAxes,
             wrap=True)
    axA.set_title("Severity", color="white", fontsize=11,
                  fontweight="bold", pad=4)

    # ── Panel B: Damage gauge ─────────────────────────────────────
    axB = fig.add_subplot(gs[1,1])
    dark_panel(axB)
    axB.text(0.5,0.87,"Surface damage", ha="center", fontsize=9,
             color="#888888", transform=axB.transAxes)
    axB.text(0.5,0.64,f"{dmg_pct:.2f}%", ha="center", fontsize=32,
             fontweight="bold", color=sc_rgb, transform=axB.transAxes)
    axB.text(0.5,0.47,"of image area is damaged", ha="center",
             fontsize=9, color="#AAAAAA", transform=axB.transAxes)
    # Progress bar background
    axB.add_patch(FancyBboxPatch((0.10,0.30),0.80,0.09,
        boxstyle="round,pad=0.01", facecolor="#2A2D3A",
        edgecolor="#444444", linewidth=0.7))
    # Progress bar fill
    fw = min(0.80, 0.80*dmg_pct/100.0)
    if fw > 0.005:
        axB.add_patch(FancyBboxPatch((0.10,0.30),fw,0.09,
            boxstyle="round,pad=0.01",
            facecolor=sc_rgb, edgecolor="none"))
    axB.text(0.10,0.22,"0%", ha="left", fontsize=7,
             color="#666666", transform=axB.transAxes)
    axB.text(0.90,0.22,"100%", ha="right", fontsize=7,
             color="#666666", transform=axB.transAxes)
    axB.text(0.5,0.12,
             f"{r['mask'].sum():,} px  /  {r['h0']*r['w0']:,} px total",
             ha="center", fontsize=8, color="#666666",
             transform=axB.transAxes)
    axB.set_title("Damage Area", color="white", fontsize=11,
                  fontweight="bold", pad=4)

    # ── Panel C: Class probabilities ─────────────────────────────
    axC = fig.add_subplot(gs[1,2])
    dark_panel(axC)
    axC.text(0.5,0.91,"Class probabilities", ha="center", fontsize=9,
             color="#888888", transform=axC.transAxes)
    cls_colors = ["#43A047","#FFB300","#E53935"]
    cls_names  = ["Minor","Moderate","Severe"]
    for k,(cn,cp,cc) in enumerate(zip(cls_names, sev_probs, cls_colors)):
        yc       = 0.73 - k*0.23
        is_pred  = (k == sev_id)
        cc_rgb   = tuple(int(cc.lstrip("#")[i:i+2],16)/255 for i in (0,2,4))
        # Highlight box for predicted class
        if is_pred:
            axC.add_patch(FancyBboxPatch((0.09,yc-0.07),0.82,0.20,
                boxstyle="round,pad=0.01", facecolor="none",
                edgecolor=cc_rgb, linewidth=1.5))
        axC.text(0.13,yc+0.06, cn, ha="left", fontsize=11,
                 color="white" if is_pred else "#888888",
                 fontweight="bold" if is_pred else "normal",
                 transform=axC.transAxes)
        axC.text(0.87,yc+0.06,f"{cp*100:.1f}%", ha="right", fontsize=11,
                 color=cc_rgb if is_pred else "#666666",
                 fontweight="bold" if is_pred else "normal",
                 transform=axC.transAxes)
        # Bar background
        axC.add_patch(FancyBboxPatch((0.13,yc-0.04),0.74,0.08,
            boxstyle="round,pad=0.01", facecolor="#2A2D3A",
            edgecolor="#444444", linewidth=0.5))
        # Bar fill
        bw = max(0.005, cp*0.74)
        axC.add_patch(FancyBboxPatch((0.13,yc-0.04),bw,0.08,
            boxstyle="round,pad=0.01", facecolor=cc_rgb, edgecolor="none",
            alpha=1.0 if is_pred else 0.35))
    axC.set_title("Class Probabilities", color="white", fontsize=11,
                  fontweight="bold", pad=4)

    # ── Panel D: Detection summary ───────────────────────────────
    axD = fig.add_subplot(gs[1,3])
    dark_panel(axD)
    axD.text(0.5,0.91,"Detection summary", ha="center", fontsize=9,
             color="#888888", transform=axD.transAxes)
    rows = [
        ("Potholes found",  str(n_holes),           sc_rgb),
        ("Damage area",     f"{dmg_pct:.2f}%",      sc_rgb),
        ("Severity",        sev_name,               sc_rgb),
        ("Confidence",      f"{sev_conf:.1f}%",     (0.8,0.8,0.8)),
        ("Model",           "ARSFormer-v3",         (0.6,0.6,0.6)),
        ("Encoder",         "MiT-B2 + FPN",         (0.6,0.6,0.6)),
    ]
    for k,(key,val,vc) in enumerate(rows):
        y = 0.79 - k*0.135
        axD.text(0.12,y,key+":", ha="left", fontsize=9,
                 color="#888888", transform=axD.transAxes)
        axD.text(0.88,y,val, ha="right", fontsize=9,
                 fontweight="bold", color=vc, transform=axD.transAxes)
        if k < len(rows)-1:
            axD.plot([0.10,0.90],[y-0.055,y-0.055],
                     color="#2A2D3A", linewidth=0.8,
                     transform=axD.transAxes)
    axD.set_title("Detection Summary", color="white", fontsize=11,
                  fontweight="bold", pad=4)

    out_path = "ARSFormer/inference_dashboard.png"
    plt.savefig(out_path, dpi=150, bbox_inches="tight",
                facecolor=fig.get_facecolor())
    plt.show()
    print(f"\n✅ Dashboard saved → {out_path}")
    return out_path


# ═══════════════════════════════════════════════════════════════════
#  UPLOAD WIDGET
# ═══════════════════════════════════════════════════════════════════

print("\n" + "="*55)
print("  ARSFormer-v3 — Pothole Detection & Severity Analysis")
print("="*55)
print("  Upload a road image below to analyse it.\n")

# File upload widget
upload_btn = widgets.FileUpload(
    accept=".jpg,.jpeg,.png,.JPG,.JPEG,.PNG",
    multiple=False,
    description="Upload Image",
    style={"button_color": "#2563eb"},
    layout=widgets.Layout(width="250px")
)

status_label = widgets.Label(
    value="⬆  Click above to upload a road image",
    layout=widgets.Layout(margin="8px 0 0 0")
)

output_area = widgets.Output()

def on_upload(change):
    if not upload_btn.value:
        return
    status_label.value = "⏳  Processing..."
    output_area.clear_output()

    try:
        # Get uploaded file bytes
        uploaded_file = list(upload_btn.value.values())[0]
        fname         = list(upload_btn.value.keys())[0]
        img_bytes     = uploaded_file["content"]

        with output_area:
            print(f"Analysing: {fname}")
            result = run_inference(bytes(img_bytes))
            clear_output(wait=True)
            plot_dashboard(result, fname=fname)

            print(f"\n{'─'*45}")
            print(f"  Severity  : {result['sev_name']}  "
                  f"({result['sev_conf']:.1f}% confidence)")
            print(f"  Damage    : {result['dmg_pct']:.2f}%  "
                  f"of image")
            print(f"  Potholes  : {result['n_holes']} detected")
            print(f"  Action    : {SEV_ACTIONS[result['sev_id']]}")
            print(f"{'─'*45}")

        status_label.value = (f"✅  Done — {result['sev_name']} severity  |  "
                              f"{result['dmg_pct']:.2f}% damage")
    except Exception as e:
        status_label.value = f"❌  Error: {e}"
        with output_area:
            print(f"Error: {e}")

upload_btn.observe(on_upload, names="value")

# Display the widget
display(widgets.VBox([
    widgets.HBox([upload_btn, status_label]),
    output_area
]))
```
